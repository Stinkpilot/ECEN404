```python
import os, shutil

# The path to the directory where the original
# dataset was uncompressed
bear_dataset_dir = '/Users/Josh/DeepLearningData/bear'
deer_dataset_dir = '/Users/Josh/DeepLearningData/deer'
wolf_dataset_dir = '/Users/Josh/DeepLearningData/wolf'
lynx_dataset_dir = '/Users/Josh/DeepLearningData/lynx2'

# The directory where we will
# store our smaller dataset
base_dir = '/Users/Josh/DeepLearningData/fourAnimalClasses'
os.mkdir(base_dir)

# Directories for our training,
# validation and test splits
train_dir = os.path.join(base_dir, 'train')
os.mkdir(train_dir)
validation_dir = os.path.join(base_dir, 'validation')
os.mkdir(validation_dir)
test_dir = os.path.join(base_dir, 'test')
os.mkdir(test_dir)

# Directory with our training cat pictures
train_bears_dir = os.path.join(train_dir, 'bears')
os.mkdir(train_bears_dir)

# Directory with our training dog pictures
train_deer_dir = os.path.join(train_dir, 'deer')
os.mkdir(train_deer_dir)

train_wolves_dir = os.path.join(train_dir, 'wolves')
os.mkdir(train_wolves_dir)

train_lynx_dir = os.path.join(train_dir, 'lynx')
os.mkdir(train_lynx_dir)

# Directory with our validation cat pictures
validation_bears_dir = os.path.join(validation_dir, 'bears')
os.mkdir(validation_bears_dir)

# Directory with our validation dog pictures
validation_deer_dir = os.path.join(validation_dir, 'deer')
os.mkdir(validation_deer_dir)

validation_wolves_dir = os.path.join(validation_dir, 'wolves')
os.mkdir(validation_wolves_dir)

validation_lynx_dir = os.path.join(validation_dir, 'lynx')
os.mkdir(validation_lynx_dir)

# Directory with our validation cat pictures
test_bears_dir = os.path.join(test_dir, 'bears')
os.mkdir(test_bears_dir)

# Directory with our validation dog pictures
test_deer_dir = os.path.join(test_dir, 'deer')
os.mkdir(test_deer_dir)

test_wolves_dir = os.path.join(test_dir, 'wolves')
os.mkdir(test_wolves_dir)

test_lynx_dir = os.path.join(test_dir, 'lynx')
os.mkdir(test_lynx_dir)


# Copy first 1500 cat images to train_cats_dir
fnames = ['bear.{}.jpg'.format(i) for i in range(1500)]
for fname in fnames:
    src = os.path.join(bear_dataset_dir, fname)
    dst = os.path.join(train_bears_dir, fname)
    shutil.copyfile(src, dst)

# Copy next 1000 cat images to validation_cats_dir
fnames = ['bear.{}.jpg'.format(i) for i in range(1500, 2000)]
for fname in fnames:
    src = os.path.join(bear_dataset_dir, fname)
    dst = os.path.join(validation_bears_dir, fname)
    shutil.copyfile(src, dst)
    
# Copy next 1000 cat images to test_cats_dir
fnames = ['bear.{}.jpg'.format(i) for i in range(2000, 2500)]
for fname in fnames:
    src = os.path.join(bear_dataset_dir, fname)
    dst = os.path.join(test_bears_dir, fname)
    shutil.copyfile(src, dst)
    
# Copy first 1000 dog images to train_dogs_dir
fnames = ['deer.{}.jpg'.format(i) for i in range(1500)]
for fname in fnames:
    src = os.path.join(deer_dataset_dir, fname)
    dst = os.path.join(train_deer_dir, fname)
    shutil.copyfile(src, dst)
    
fnames = ['lynx.{}.jpg'.format(i) for i in range(1500)]
for fname in fnames:
    src = os.path.join(lynx_dataset_dir, fname)
    dst = os.path.join(train_lynx_dir, fname)
    shutil.copyfile(src, dst)

fnames = ['wolf.{}.jpg'.format(i) for i in range(1500)]
for fname in fnames:
    src = os.path.join(wolf_dataset_dir, fname)
    dst = os.path.join(train_wolves_dir, fname)
    shutil.copyfile(src, dst)
# Copy next 500 dog images to validation_dogs_dir
fnames = ['deer.{}.jpg'.format(i) for i in range(1500, 2000)]
for fname in fnames:
    src = os.path.join(deer_dataset_dir, fname)
    dst = os.path.join(validation_deer_dir, fname)
    shutil.copyfile(src, dst)
    
fnames = ['lynx.{}.jpg'.format(i) for i in range(1500, 2000)]
for fname in fnames:
    src = os.path.join(lynx_dataset_dir, fname)
    dst = os.path.join(validation_lynx_dir, fname)
    shutil.copyfile(src, dst)

fnames = ['wolf.{}.jpg'.format(i) for i in range(1500, 2000)]
for fname in fnames:
    src = os.path.join(wolf_dataset_dir, fname)
    dst = os.path.join(validation_wolves_dir, fname)
    shutil.copyfile(src, dst)
# Copy next 500 dog images to test_dogs_dir
fnames = ['deer.{}.jpg'.format(i) for i in range(2000, 2500)]
for fname in fnames:
    src = os.path.join(deer_dataset_dir, fname)
    dst = os.path.join(test_deer_dir, fname)
    shutil.copyfile(src, dst)
    
fnames = ['lynx.{}.jpg'.format(i) for i in range(2000, 2500)]
for fname in fnames:
    src = os.path.join(lynx_dataset_dir, fname)
    dst = os.path.join(test_lynx_dir, fname)
    shutil.copyfile(src, dst)  
    
fnames = ['wolf.{}.jpg'.format(i) for i in range(2000, 2500)]
for fname in fnames:
    src = os.path.join(wolf_dataset_dir, fname)
    dst = os.path.join(test_wolves_dir, fname)
    shutil.copyfile(src, dst)   
```
